### Création des Dataframes

- On créé un df par dossier. Un df est la concaténation de l'ensemble des fichiers csv à l'intérieur de chaque dossier

```python

# Chemin vers le répertoire contenant les sous-dossiers avec les fichiers CSV
root_folder_path = 'C:/Users/PELLE Mayel/Documents/Perso/LOTO/stream_stat/data/'

# Dictionnaire pour stocker les noms des DataFrames par dossier
dfs_dict = {}

# Parcourir chaque dossier dans le répertoire racine
for folder in os.listdir(root_folder_path):
    folder_path = os.path.join(root_folder_path, folder)
    
    # Vérifier si c'est un dossier
    if os.path.isdir(folder_path):
        # Liste pour stocker les DataFrames individuels
        dfs = []
        
        # Parcourir chaque fichier CSV dans le dossier
        for file in os.listdir(folder_path):
            if file.endswith('.csv'):
                file_path = os.path.join(folder_path, file)
                df = pd.read_csv(file_path, sep=";")
                dfs.append(df)
        
        # Concatenation des DataFrames en un seul DataFrame pour le dossier
        combined_df = pd.concat(dfs, ignore_index=True)
        
        # Nommer dynamiquement le DataFrame
        df_name = f'df_{folder}'
        exec(f"{df_name} = combined_df")
        
        # Stocker le DataFrame dans le dictionnaire
        dfs_dict[df_name] = combined_df

# Afficher les noms et le contenu des DataFrames par dossier
for df_name, df in dfs_dict.items():
    print(f"DataFrame: {df_name}")
    print(df)
```
